```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
    sns.heatmap(df.corr())
```

```python
#People with heart attack seems to be mostly between ages 40 to 55 
sns.kdeplot(df[df['target']==1]['age'],color="RED")
sns.kdeplot(df[df['target']==0]['age'],color="BLUE")
```

```python
#Serum cholestrol doesn't contribute to heart attack
sns.distplot(df[df['target']==1]['chol'],color="RED")
sns.distplot(df[df['target']==0]['chol'],color="BLUE")
```

```python
#Chest pain is seen along with heart attack
sns.distplot(df[df['target']==1]['cp'],color="RED")
sns.distplot(df[df['target']==0]['cp'],color="BLUE")
```

```python
sns.kdeplot(
    data=df,
    x="chol",
    y="thalach",
    hue="target",
    thresh=.1,
)
```
